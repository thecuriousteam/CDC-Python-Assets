```python
# strings

info = "Curious Developers Community"
print(type(info))

# accessing
print(info[0:7])
print(info[:])

# looping
for char in range(len(info)):
  print(info[char], end =" ")

print()
for char in info:
  print(char, end =" ")

print()  
for char in info[0:10:2]:
  print(char, end =" ")

print()
print(info.find('D'))
print(info.__contains__('D'))
print(info.__len__())
print(info.__eq__('Curious Developers Community'))
print(info.__getitem__(8))
print(info.count('e'))
```

```python
# list
info =["Curious","Developers" , "Community", "A", "Open Source", "Community"]
print(type(info))

# accessing the list
print(info[0:1])
print(info[:-1])
print(info[0:1:2])

# looping
for data in info:
  print(data)
  
for data in range(len(info)):
  print(info[data])
  
# some builtin methods
# adding
info.append("Curious Ecosystem")
print("Final data: ", info)

# soring
info.sort()
print("Sorting the list: ",info)
print("Checking the presence of 'Curious Ecosystem' in a list: ",info.__contains__('Curious Ecosystem'))

# updating
info[1] = "DEVELOPERS"
print("After updating the list: ",info)

# removing
info.remove('Community')
print("After removing the element: ",info)

# reversing
info.reverse()
print("After reversing the list: ",info)

# taking data from the tuples and appending to list
data = ('Product developemnt','Consultant','Design')
for eachData in data:
  info.append(eachData)
print("Final data after adding records: ",info)
```

```python
info =("Curious","Developers" , "Community", "A", "Open Source", "Community")
print(type(info))

# accessing the tuple
print(info[0:1])
print(info[:-1])
print(info[0:1:2])

# looping

for data in info:
  print(data)
for eachData in range(len(info)):
  print(info[eachData], end=" ")
  
print()

# appending
"""
1. Convert to list
2. addpend the data to the list
3. convert the appendied data to a tuple
"""
info = list(info)
info.append('Curious Ecosystem')
info = tuple(info)
print("After appending the new data: ",info)

# updating
"""
1. Convert to list
2. update the data to the list
3. convert the appendied data to a tuple
"""
info = list(info)
info[1] = "AAA"
info = tuple(info)
print("After updating the new data: ",info)

# removing
info = list(info)
info.remove('AAA')
info = tuple(info)
print("Fianl tuple after removing: ",info)


info =("Curious","Developers" , "Community", "A", "Open Source", "Community")
print(info)
```

```python
# dictionary
data = {1:"One",2:"Two",3:"Three",4:"Four",5:"Five"}
print(data)

# accessing - getting value based on key
print(data.get(1)) 
print(data.get(2))
print("Getting keys: ",data.keys())
print("Getting values: ",data.values())
print("Size of dictionary: ",data.__len__())

# looping
for keys,values in data.items():
  print("Key : {keys} and Value : {values}".format(keys=keys,values=values))
  
for keys in data.keys():
  print("Key : {keys}".format(keys=keys))
  
for values in data.values():
  print("Values : {values}".format(values=values))
  
# forming dictionary from 2 lists
keyList = ['a','b','c','d','e','f']
valueList = ['aaa','bbb','ccc','ddd','eee','fff']

# way - 1
newDict = dict(zip(keyList, valueList))
print("Final dictionary information: ",newDict)

# way - 2
newFinalDict = {}
for i,j in enumerate(keyList):
  newFinalDict[keyList[i]] = valueList[i]
print("Final dictionary information: ",newFinalDict)

data = {1:"One",2:"Two",3:"Three",4:"Four",5:"Five"}

# adding element
data[6] = "SIX"


# updating elements
data.update({1:"ONE ONE"})
print(data)
```
